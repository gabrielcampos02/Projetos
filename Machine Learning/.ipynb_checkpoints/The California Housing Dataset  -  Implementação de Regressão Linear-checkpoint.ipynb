### Multicolinearidade

```python
# Avaliação da Multicolinearidade - Matriz de Correlação

matriz_corr = X.corr()
matriz_corr
```

```python
def visualize_correlation_matrix(data, hurdle = 0.0):
    R = np.corrcoef(data, rowvar = 0)
    R[np.where(np.abs(R) < hurdle)] = 0.0
    heatmap = plt.pcolor(R, cmap = mpl.cm.coolwarm, alpha = 0.8)
    heatmap.axes.set_frame_on(False)
    heatmap.axes.set_yticks(np.arange(R.shape[0]) + 0.5, minor = False)
    heatmap.axes.set_xticks(np.arange(R.shape[1]) + 0.5, minor = False)
    heatmap.axes.set_xticklabels(variables, minor = False)
    plt.xticks(rotation=90)
    heatmap.axes.set_yticklabels(variables, minor = False)
    plt.tick_params(axis = 'both', which = 'both', bottom = 'off', top = 'off', left = 'off', right = 'off') 
    plt.colorbar()
    plt.show()
```

```python
visualize_correlation_matrix(X, hurdle = 0.5)
```

--------------------------------------------------------------------------------------------------------------------------------
__Obs:__ Fazendo uma análise tanto na matriz de correlação, como no mapa de calor, nota-se uma correlação negativa forte entre as variáveis __"Latitude"__ e __"Longitude"__, isso por conta das características geográficas do Estado da California que tem uma tendencia de queda diagonal, com a latitude aumentando em direção ao Norte e com a Longitude diminuindo em relação ao Oeste. Outra correlação forte, desta vez positiva, está entre as variáveis __"AveBedrms"__ e __"AveRooms"__, o que possivelmente indica que manter essas duas variáveis juntas no modelo pode gerar redundância.

### Gradiente Descendente

```python
# Gerando os dados para aplicação do Gradiente Descendente

observations = len(dataset)
variables = dataset.columns
```

```python
# Aplicando Padronização aos dados

standardization = StandardScaler()
Xst = standardization.fit_transform(X)
original_means = standardization.mean_
originanal_stds = standardization.scale_
```

--------------------------------------------------------------------------------------------------------------------------------
__Obs:__ Para poder aplicar o gradiente descendente de forma correta é necessário que os dados estejam na mesma escala, para isso foi aplicado uma técnica de padronização para os mesmos.

```python
# Gerando X e Y 

Xst = np.column_stack((Xst,np.ones(observations)))
y  = dataset['target'].values
```

```python
# Implementação do algoritmo de Gradiente Descendente

import random
import numpy as np

def random_w( p ):
    return np.array([np.random.normal() for j in range(p)])

def hypothesis(X,w):
    return np.dot(X,w)

def loss(X,w,y):
    return hypothesis(X,w) - y

def squared_loss(X,w,y):
    return loss(X,w,y)**2

def gradient(X,w,y):
    gradients = list()
    n = float(len( y ))
    for j in range(len(w)):
        gradients.append(np.sum(loss(X,w,y) * X[:,j]) / n)
    return gradients

def update(X,w,y, alpha = 0.01):
    return [t - alpha*g for t, g in zip(w, gradient(X,w,y))]

def optimize(X,y, alpha = 0.01, eta = 10**-12, iterations = 1000):
    w = random_w(X.shape[1])
    path = list()
    for k in range(iterations):
        SSL = np.sum(squared_loss(X,w,y))
        new_w = update(X,w,y, alpha = alpha)
        new_SSL = np.sum(squared_loss(X,new_w,y))
        w = new_w
        if k>=5 and (new_SSL - SSL <= eta and new_SSL - SSL >= -eta):
            path.append(new_SSL)
            return w, path
        if k % (iterations / 20) == 0:
            path.append(new_SSL)
    return w, path 
```

```python
# Gerando os coeficientes finais padronizados

alpha = 0.01
w, path = optimize(Xst, y, alpha, eta = 10**-12, iterations = 200000)
print ("Coeficientes finais padronizados: " + ', '.join(map(lambda x: "%0.4f" % x, w)))  
```

```python
# Desfazendo a Padronização

unstandardized_betas = w[:-1] / originanal_stds
unstandardized_bias  = w[-1]-np.sum((original_means / originanal_stds) * w[:-1])
```

```python
# Coeficientes Finais

print ('%8s: %8.4f' % ('bias', unstandardized_bias))
for beta,varname in zip(unstandardized_betas, variables):
    print ('%8s: %8.4f' % (varname, beta))
```

--------------------------------------------------------------------------------------------------------------------------------
__Obs:__ Com a aplicação do gradiente descendente é possivel minimizar a função de perda e assim otimizar o modelo.

### Calculo do R-Squared

```python
# Criando o modelo

modelo = linear_model.LinearRegression(fit_intercept = True)
```

```python
# Calculo do R-Squared

def r2_est(X,y):
    return r2_score(y, modelo.fit(X,y).predict(X))

print ('Coeficiente R2: %0.3f' %  r2_est(X,y))
```

```python
# Calculo do impacto de R-Squared por variável

r2_impact = list()
for j in range(X.shape[1]):
    selection = [i for i in range(X.shape[1]) if i!=j]
    r2_impact.append(((r2_est(X,y) - r2_est(X.values[:,selection],y)), dataset.columns[j]))
    
for imp, varname in sorted(r2_impact, reverse = True):
    print ('%6.3f %s' %  (imp, varname))
```

--------------------------------------------------------------------------------------------------------------------------------
__Obs__: Fazendo uma análise do resultado é possivel observar o impacto de cada variável tem na contrução do R2 para o modelo, isso também possui uma coerência pois __"MedInc"__, __"Latitude"__ e __"Longitude"__, são as variáveis que geram maior impacto na contrução do R2, corroborando com o fato de serem as variáveis mais relevantes para o modelo em si.
